```python
import pandas as pd
```

```python
Working_Woman = [('Shreya', 28, 'Himachal'),
            ('Sanju', 32, 'Delhi'),
            ('Alia', 29, 'Mumbai'),
            ('Sanjana', 25, 'Delhi'),
            ('Saumya', 25, 'Delhi'),
            ('Sanjana', 30, 'Delhi'),
            ('Sanjana', 25, 'Dehradun'),
            ('Rinki', 32, 'Kolkata'),
            ('Sanju', 32, 'Delhi'),
            ('Seema', 32, 'Delhi'),
            ('Rinki', 30, 'Delhi'),
            ('Sanju', 32, 'Delhi'),
            ('Ramya', 28, 'Bangalore')
            ]
```

```python
df = pd.DataFrame(Working_Woman, columns = ['Name', 'Age', 'City'])
```
